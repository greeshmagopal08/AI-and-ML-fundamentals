# Question 4

Initially when the PCA is performed, the feature selection is done by removing each feature based on the minimum covariance. In this case, the feature is not actually removed but it exists in a different form(transpose). So when the feature with minimum covariance is eliminated there is not much difference in the accuracy(mean square error).
When the backward search was done, covariance was not considered while removing the features. The feature was completely removed and in this case there was huge error because the feature removed is impacting the data.
