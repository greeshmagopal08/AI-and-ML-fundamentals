```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix
import math
```

```python
data_test=pd.read_excel ('/Users/anandramesh/Desktop/ML/ML Assignment 2/test.xlsx')         
data_train=pd.read_excel ('/Users/anandramesh/Desktop/ML/ML Assignment 2/train.xlsx')
data_test
```

### Label encoding the given test and train data

```python
data_test['University_state']=data_test['University_state'].map({'New York':0,'Massachusetts':1,'California':2})
data_test['Target']=data_test['Target'].map({'public':0,'private':1})
data_train['University_state']=data_train['University_state'].map({'New York':0,'Massachusetts':1,'California':2})
data_train['Target']=data_train['Target'].map({'public':0,'private':1})
```

### Splitting the data to zeros/ones and test/train data

```python
Xtrain=data_train.drop(['University_name','Target'],axis=1)
Xtest=data_test.drop(['University_name','Target'],axis=1)
Ytrain=data_train.Target
Ytest=data_test.Target
```

```python
Xtrain.shape
```

#### Seperate zeros and ones 

```python
data_new=pd.read_excel('/Users/anandramesh/Desktop/ML/ML Assignment 2/train.xlsx')
data_new.Target=data_new.Target.map({'public':0,'private':1})
data_new.University_state=data_new.University_state.map({'New York':1,'Massachusetts':2,'California':3})
```

#### Calculating mean and standard deviation for test and train data

```python
#train data
zeros_train=data_train[data_train['Target']==0]
ones_train= data_train[data_train['Target']==1]
ones_stdmean=ones_train.describe()
zeros_stdmean=zeros_train.describe()
zeros_stdmean
```

### Training data probability calculation 

```python
Xtrain.shape[0]
```

```python
def nb_probability(x, mean, stdev):
    e = math.exp(-((x-mean)**2 / (2 * stdev**2 )))
    return (1 / (math.sqrt(2 * math.pi) * stdev)) * e
```

```python
#Adding the probability calculated into the dataframe
Final_data_train=pd.read_excel('/Users/anandramesh/Desktop/ML/ML Assignment 2/train.xlsx')
Final_data_train['probofzero']=np.nan
Final_data_train['probofone']=np.nan
```

```python
Final_data_test=pd.read_excel('/Users/anandramesh/Desktop/ML/ML Assignment 2/test.xlsx')
```

```python
Final_data_train
```

```python
Final_data_test
```

```python
#Considering the training data
trainprob_1=[]
trainprob_0=[]
for i in ['University_state','SAT_verbal','SAT_math','Academics','Social','Quality_of_life']:
    prob_1=[]
    prob_0=[]
    for j in range(Xtrain.shape[0]):
        ones_prob=nb_probability(Xtrain[i].iloc[j],ones_stdmean.loc['mean',i],ones_stdmean.loc['std',i])
        zeros_prob=nb_probability(Xtrain[i].iloc[j],zeros_stdmean.loc['mean',i],zeros_stdmean.loc['std',i])
        prob_1.append(ones_prob)
        prob_0.append(zeros_prob)
    trainprob_1.append(prob_1)
    trainprob_0.append(prob_0)
resultone=pd.DataFrame(columns=['Univstate1','SATVerb1','SATMath1','Academics1','Social1','Qualityoflife1'])
resultzero=pd.DataFrame(columns=['Univstate0','SATVerb0','SATMath0','Academics0','Social0','Qualityoflife0'])
```

```python
for i in range(len(resultone.columns)):
    resultone.iloc[:,i]=trainprob_1[i]
    resultzero.iloc[:,i]=trainprob_0[i]
Final_data_train['probofone']=resultone['Univstate1']*resultone['SATVerb1']*resultone['SATMath1']*resultone['Academics1']*resultone['Social1']*resultone['Qualityoflife1']
Final_data_train['probofzero']=resultzero['Univstate0']*resultzero['SATVerb0']*resultzero['SATMath0']*resultzero['Academics0']*resultzero['Social0']*resultzero['Qualityoflife0']
```

```python
#Considering the one with higher probability
Final_data_train['Class']=Final_data_train.probofone>Final_data_train.probofzero
Final_data_train.Target=Final_data_train.Target.map({'public':0,'private':1})
Final_data_train.Class=Final_data_train.Class.map({False:0,True:1})
```

```python
print('Accuracy score of the train data using NB:', accuracy_score(Ytrain,Final_data_train.Class)*100)
print('Confusion matrix for the train data using NB:',confusion_matrix(Ytrain,Final_data_train.Class))
```

```python
#Probability for the test data
testprob_1=[]
testprob_0=[]
for i in ['University_state','SAT_verbal','SAT_math','Academics','Social','Quality_of_life']:
    prob_1=[]
    prob_0=[]
    for j in range(Xtest.shape[0]):
        ones_prob=nb_probability(Xtest[i].iloc[j],ones_stdmean.loc['mean',i],ones_stdmean.loc['std',i])
        zeros_prob=nb_probability(Xtest[i].iloc[j],zeros_stdmean.loc['mean',i],zeros_stdmean.loc['std',i])
        prob_1.append(ones_prob)
        prob_0.append(zeros_prob)
    testprob_1.append(prob_1)
    testprob_0.append(prob_0)
resultone=pd.DataFrame(columns=['Univstate1','SATVerb1','SATMath1','Academics1','Social1','Qualityoflife1'])
resultzero=pd.DataFrame(columns=['Univstate0','SATVerb0','SATMath0','Academics0','Social0','Qualityoflife0'])
```

```python
for i in range(len(resultone.columns)):
    resultone.iloc[:,i]=testprob_1[i]
    resultzero.iloc[:,i]=testprob_0[i]
Final_data_test['probofone']=resultone['Univstate1']*resultone['SATVerb1']*resultone['SATMath1']*resultone['Academics1']*resultone['Social1']*resultone['Qualityoflife1']
Final_data_test['probofzero']=resultzero['Univstate0']*resultzero['SATVerb0']*resultzero['SATMath0']*resultzero['Academics0']*resultzero['Social0']*resultzero['Qualityoflife0']
```

```python
#Considering the one with higher probability
Final_data_test['Class']=Final_data_train.probofone>Final_data_train.probofzero
Final_data_test.Target=Final_data_test.Target.map({'public':0,'private':1})
Final_data_test.Class=Final_data_test.Class.map({False:0,True:1})
```
